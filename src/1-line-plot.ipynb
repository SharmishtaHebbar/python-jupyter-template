```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a 5x5 matrix with random numbers (0-1) using a fixed seed
np.random.seed(42)
matrix = np.random.rand(5, 5)

# Step 2: Compute inverse and handle non-invertible matrices
try:
    matrix_inv = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    matrix_inv = None
    print("Matrix is non-invertible.")

# Step 3: Verify closeness to identity matrix if invertible
if matrix_inv is not None:
    identity_check = np.dot(matrix, matrix_inv)
    print("\nIdentity Check (Original * Inverse):\n", np.round(identity_check, 2))

# Step 4: Statistical Analysis
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)

# Display results
print("\nOriginal Matrix:")
print(np.round(matrix, 2))
print("\nStatistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")

# Step 5: Create a line plot
x = np.arange(0, 11)
y = x ** 2

# Error handling for input data
if x is None or y is None or len(x) == 0 or len(y) == 0 or len(x) != len(y):
    raise ValueError("Invalid input data for plotting.")

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Square of X")
plt.xlabel("X Values")
plt.ylabel("Y = X^2")
plt.grid(True)
plt.show()
```
