```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a 5x5 matrix with random numbers (0-1) using a fixed seed
np.random.seed(42)
matrix = np.random.rand(5, 5)

# Step 2: Compute inverse and handle non-invertible matrices
try:
    matrix_inv = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    matrix_inv = None
    print("Error: Matrix is non-invertible.")

# Step 3: Verify closeness to identity matrix if invertible
if matrix_inv is not None:
    identity_check = np.dot(matrix, matrix_inv)
    print("\nIdentity Check (Original * Inverse):\n", np.round(identity_check, 2))

# Step 4: Statistical Analysis
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)

# Display results
print("\nOriginal Matrix:")
print(np.round(matrix, 2))
print("\nStatistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")

# Step 5: Create a line plot with error handling
try:
    x = np.linspace(0, 10, 100)
    y = x ** 2
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='y = x^2', color='b')
    plt.title("Line Plot of x vs x^2")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.legend()
    plt.show()
except Exception as e:
    print(f"Error in line plot: {e}")

# Step 6: Create a bar plot with error handling
try:
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['red', 'blue', 'green', 'purple']
    plt.figure(figsize=(8, 5))
    plt.bar(categories, values, color=colors)
    plt.title("Bar Plot of Categories")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
except Exception as e:
    print(f"Error in bar plot: {e}")

# Step 7: Create a scatter plot with annotations and error handling
try:
    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    plt.figure(figsize=(8, 5))
    plt.scatter(x_scatter, y_scatter, color='blue')
    for i in range(len(x_scatter)):
        plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.show()
except Exception as e:
    print(f"Error in scatter plot: {e}")

# Step 8: Create a pie chart with error handling
try:
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    plt.figure(figsize=(8, 5))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'blue', 'red', 'green'], startangle=140)
    plt.title("Programming Language Popularity")
    plt.show()
except Exception as e:
    print(f"Error in pie chart: {e}")

# Step 9: Create a histogram with error handling
try:
    hist_data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
    plt.figure(figsize=(8, 5))
    plt.hist(hist_data, bins=5, color='purple', edgecolor='black')
    plt.title("Histogram of Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
except Exception as e:
    print(f"Error in histogram: {e}")
```
