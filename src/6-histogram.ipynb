```python
import matplotlib.pyplot as plt
import numpy as np

def validate_data(data):
    """Checks if data is a non-empty list of numbers."""
    if not isinstance(data, list):
        raise ValueError("Input data must be a list.")
    if not data:
        raise ValueError("Input data cannot be empty.")
    if not all(isinstance(x, (int, float)) for x in data):
        raise ValueError("All elements in the data list must be numbers.")
    return True

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

try:
    # Validate data before processing
    validate_data(data)
    
    # Create histogram
    plt.hist(data, bins=np.arange(min(data), max(data) + 2) - 0.5, color='blue', edgecolor='black')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram of Given Data')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Show plot
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
```
