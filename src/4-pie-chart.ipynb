### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a 5x5 matrix with random numbers (0-1) using a fixed seed
np.random.seed(42)
matrix = np.random.rand(5, 5)

# Step 2: Compute inverse and handle non-invertible matrices
try:
    matrix_inv = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    matrix_inv = None
    print("Matrix is non-invertible.")

# Step 3: Verify closeness to identity matrix if invertible
if matrix_inv is not None:
    identity_check = np.dot(matrix, matrix_inv)
    print("\nIdentity Check (Original * Inverse):\n", np.round(identity_check, 2))

# Step 4: Statistical Analysis
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)

# Display results
print("\nOriginal Matrix:")
print(np.round(matrix, 2))
print("\nStatistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")

# Step 5: Create a line plot
x = np.linspace(0, 10, 100)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='b')
plt.title("Line Plot of x vs x^2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.legend()
plt.show()

# Step 6: Create a bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.title("Bar Plot of Categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 7: Create a scatter plot with annotations
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='red', label='Data Points')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(f"{txt}", (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Create a pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'], startangle=140)
plt.title("Pie Chart of Programming Languages")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def validate_data(data, expected_length=None):
    """Check if data is valid and meets expected criteria."""
    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Input data must be a list or numpy array.")
    if expected_length and len(data) != expected_length:
        raise ValueError(f"Input data must have exactly {expected_length} elements.")
    if any(d is None for d in data):
        raise ValueError("Input data contains None values.")

# Data for subplots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Validate input data
validate_data(x)
validate_data(y)
validate_data(categories, 4)
validate_data(values, 4)
validate_data(labels, 4)
validate_data(sizes, 4)

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axes[0, 0].plot(x, y, marker='o', linestyle='-')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")
axes[0, 0].grid(True)

# Bar chart
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[0, 1].set_title("Bar Chart")

# Scatter plot
axes[1, 0].scatter(x, y, color='purple')
axes[1, 0].set_title("Scatter Plot")
for i, txt in enumerate(y):
    axes[1, 0].annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'pink'])
axes[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
