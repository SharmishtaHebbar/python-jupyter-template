### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a 5x5 matrix with random numbers (0-1) using a fixed seed
np.random.seed(42)
matrix = np.random.rand(5, 5)

# Step 2: Compute inverse and handle non-invertible matrices
try:
    matrix_inv = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    matrix_inv = None
    print("Matrix is non-invertible.")

# Step 3: Verify closeness to identity matrix if invertible
if matrix_inv is not None:
    identity_check = np.dot(matrix, matrix_inv)
    print("\nIdentity Check (Original * Inverse):\n", np.round(identity_check, 2))

# Step 4: Statistical Analysis
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)

# Display results
print("\nOriginal Matrix:")
print(np.round(matrix, 2))
print("\nStatistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")

# Step 5: Create a line plot
x = np.linspace(0, 10, 100)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='b')
plt.title("Line Plot of x vs x^2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.legend()
plt.show()

# Step 6: Create a bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.title("Bar Plot of Categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 7: Create a scatter plot with annotations
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='r', label='Data Points')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(txt, (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a 5x5 matrix with random numbers (0-1) using a fixed seed
np.random.seed(42)
matrix = np.random.rand(5, 5)

# Step 2: Compute inverse and handle non-invertible matrices
try:
    matrix_inv = np.linalg.inv(matrix)
except np.linalg.LinAlgError:
    matrix_inv = None
    print("Matrix is non-invertible.")

# Step 3: Verify closeness to identity matrix if invertible
if matrix_inv is not None:
    identity_check = np.dot(matrix, matrix_inv)
    print("\nIdentity Check (Original * Inverse):\n", np.round(identity_check, 2))

# Step 4: Statistical Analysis
mean_value = np.mean(matrix)
median_value = np.median(matrix)
variance_value = np.var(matrix)

# Display results
print("\nOriginal Matrix:")
print(np.round(matrix, 2))
print("\nStatistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Variance: {variance_value:.2f}")

# Step 5: Create a line plot with x values ranging from 0 to 10 and y values as the square of x
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y, label='y = x^2')
plt.title('Line Plot of y = x^2')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()
plt.show()

# Exercise 2: Bar Plot with Colors
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
colors = ['red', 'blue', 'green', 'purple', 'orange']

# Error Handling Before Processing Input of Bar Plot
if not (isinstance(categories, list) and isinstance(values, list) and isinstance(colors, list)):
    raise ValueError("Inputs for bar plot must be lists.")
if len(categories) != len(values) or len(categories) != len(colors):
    raise ValueError("All input lists must have the same length.")

plt.bar(categories, values, color=colors)
plt.title('Bar Plot with Colors')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Exercise 3: Scatter Plot with Annotations
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Error Handling Before Processing Input of Scatter Plot
if not (isinstance(x_scatter, list) and isinstance(y_scatter, list)):
    raise ValueError("Inputs for scatter plot must be lists.")
if len(x_scatter) != len(y_scatter):
    raise ValueError("Input lists for scatter plot must have the same length.")

plt.scatter(x_scatter, y_scatter, color='blue')
for i in range(len(x_scatter)):
    plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```
